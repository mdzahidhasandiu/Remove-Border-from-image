```python
Folder = r'/content/drive/MyDrive/Python/Input/Photo'
```

```python
import os
import cv2
import numpy as np
# from matplotlib import pyplot as plt 

count = 0
for img in os.listdir(Folder):
    print(img)
    img = cv2.imread(os.path.join(Folder, img))  
    h, w = img.shape[:2] 
    print(h)
    print(w)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
    # add 1 pixel white border all around 
    pad = cv2.copyMakeBorder(gray, 1,1,1,1, cv2.BORDER_CONSTANT, value=255) 
    h, w = pad.shape 
    print(h)
    print(w)
 
    # create zeros mask 2 pixels larger in each dimension 
    mask = np.zeros([h + 2, w + 2], np.uint8)
    # floodfill outer white border with black 
    img_floodfill = cv2.floodFill(pad, mask, (0,0), 0, (5), (0), flags=8)[1]  
    # remove border 
    img_floodfill = img_floodfill[1:h-1, 1:w-1]  
    count += 1
    cv2.imwrite(f'/content/drive/MyDrive/Python/Input/Output/Image_{count}.jpg', img_floodfill)
```
